```python
def plot_multiple_time_acc(data_paths, axis):
    cp_data_paths = data_paths.copy()
    fig = pd.read_csv(cp_data_paths[0], usecols=['t', axis]).plot(x='t',y=axis)
    cp_data_paths.remove(cp_data_paths[0])
    for path in cp_data_paths:
        df = pd.read_csv(path, usecols=['t', axis])
        df.plot(x='t', y=axis, ax=fig)
    plt.show()
```

```python
def plot_multiple_time_acc_groupings(data_paths_groupings, axis):
    for data_paths in data_paths_groupings:
        plot_multiple_time_acc(data_paths, axis)
```

```python
def plot_axis_time_acc(data_path):
    df = pd.read_csv(data_path, usecols=['t', 'x','y','z', 'l'])
    df.plot(x='t',y=['x','y','z', 'l'])
    plt.xlabel('Time (ms)')
    plt.ylabel('Acceleration (g)')
    plt.legend(['X-Axis', 'Y-Axis', 'Z-Axis', 'Light'])
    plt.show()
```

```python
def plot_multiple_axis_time_acc(data_paths):
    for path in data_paths:
        plot_axis_time_acc(path)
```
