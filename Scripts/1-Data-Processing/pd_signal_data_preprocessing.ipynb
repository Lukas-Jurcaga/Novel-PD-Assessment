```python
old_dataset_dir = '../../Study/Data/labelled/'
new_dataset_dir = '../../Study/Data/cleaned/'
```

# Preprocessing PD Study Data

### Basing Time Variable

```python
record_paths = os.listdir(old_dataset_dir)
```

```python
get_light_recordings(pd.read_csv(old_dataset_dir + record_paths[0]))
```

## Trimming Records

```python
# Trimming the start
starting_indexes = []
trimmed_dfs = []
i = 0
for rec_path in record_paths:
    rec_df = pd.read_csv(old_dataset_dir + rec_path)
    light_recordings_df = get_light_recordings(rec_df)
    # Finding the start index
    first_light_recording = light_recordings_df.iloc[0]
    # Not including any partial light signals
    if not is_time_long_enough(first_light_recording, light_recordings_df.iloc[1]):
        first_light_recording = light_recordings_df.iloc[1]
    start_index = first_light_recording['start_index']
    
    # Finding the end index
    push_index_by = 0
    if len(light_recordings_df) >= 18:
        push_index_by = 1
    last_light_recording = light_recordings_df.iloc[len(light_recordings_df) - 1]
    if is_time_long_enough(last_light_recording, light_recordings_df.iloc[len(light_recordings_df) - 2]):
        last_light_recording = light_recordings_df.iloc[len(light_recordings_df) - 2 - push_index_by]
    else:
        last_light_recording = light_recordings_df.iloc[len(light_recordings_df) - 3 - push_index_by]
    end_index = last_light_recording['end_index']
    new_df = rec_df.iloc[int(start_index):int(end_index + 1)]
    trimmed_dfs.append(new_df.reset_index().drop('index', axis=1))
        
```
